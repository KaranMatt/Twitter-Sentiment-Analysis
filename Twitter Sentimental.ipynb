```python
import numpy as np
import pandas as pd

train_df=pd.read_csv('Data/twitter_training.csv')
test_df=pd.read_csv('Data/twitter_validation.csv') #test set is small (999 rows) 
train_df
```

```python
train_df.columns=['tweet_id','topic','sentiment','text']
test_df.columns=['tweet_id','topic','sentiment','text']
```

```python
combined_df=pd.concat([train_df,test_df]) # combining both to make a large dataset and then splitting 
combined_df
```

```python
train_df.info()
```

```python
combined_df.info()
```

```python
combined_df.dropna(subset=['text'],inplace=True)
```

```python
from nltk.tokenize import TweetTokenizer
tweet_token=TweetTokenizer()
combined_df['tokens']=combined_df['text'].apply(tweet_token.tokenize)
combined_df
```

```python
combined_df['lower_tokens']=combined_df['tokens'].apply(lambda tokens:[token.lower() for token in tokens])
combined_df
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
combined_df['no_stopwords']=combined_df['lower_tokens'].apply(lambda tokens:[token for token in tokens if token not in stop_words])
```

```python
combined_df
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
combined_df['lemmatized_tokens']=combined_df['no_stopwords'].apply(lambda tokens:[lemmatizer.lemmatize(token,pos='v') for token in tokens])
combined_df
```

```python
combined_df['cleaned_tokens']=combined_df['lemmatized_tokens'].apply(lambda tokens:[token for token in tokens if token.isalnum()])
combined_df
```

```python
combined_df['processed_text']=combined_df['cleaned_tokens'].apply(lambda tokens:' '.join(tokens))
combined_df
```

```python
code_map={'Positive':0,'Neutral':1,'Negative':2,'Irrelevant':3}
combined_df['code']=combined_df['sentiment'].map(code_map)
combined_df
```

```python
from sklearn.model_selection import train_test_split
train_set,val_set=train_test_split(combined_df,test_size=0.2,random_state=42,stratify=combined_df['code'])
train_set.shape,val_set.shape
```

```python
train_df['sentiment'].value_counts() #classes are almost balanced
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf=TfidfVectorizer()
train_tfidf=tf_idf.fit_transform(train_set['processed_text'])
val_tfidf=tf_idf.transform(val_set['processed_text'])
```

```python
train_label=train_set['code']
val_label=val_set['code']
```

```python
import mlflow
import mlflow.sklearn
import mlflow.keras
```

```python
from sklearn.metrics import accuracy_score,f1_score
def model_evaluate(model):
    train_preds=model.predict(train_tfidf)
    val_preds=model.predict(val_tfidf)
    train_acc=accuracy_score(train_label,train_preds)
    val_acc=accuracy_score(val_label,val_preds)
    train_f1=f1_score(train_label,train_preds,average='weighted')
    val_f1=f1_score(val_label,val_preds,average='weighted')
    print(f'Train Results: Accuracy ==> {train_acc} and F1 ==> {train_f1}')
    print(f'Test Results: Accuracy ==> {val_acc} and F1 ==> {val_f1}')
    mlflow.log_metric('Train_Accuracy',train_acc)
    mlflow.log_metric('Train_F1',train_f1)
    mlflow.log_metric('Test_Accuracy',val_acc)
    mlflow.log_metric('Test_F1',val_f1)
```

```python
mlflow.set_experiment('Twitter Sentiment Analysis')
```

```python
from sklearn.linear_model import LogisticRegression
with mlflow.start_run(run_name='Logistic Regression'):
    model_log=LogisticRegression(class_weight='balanced')
    model_log.fit(train_tfidf,train_label)
    mlflow.log_params(model_log.get_params())
    mlflow.set_tag('model_type','Logistic Regression')
    model_evaluate(model_log)
```

```python
from sklearn.tree import DecisionTreeClassifier
with mlflow.start_run(run_name='Decision Tree v3'):
    model_tree=DecisionTreeClassifier(max_depth=85,random_state=42,class_weight='balanced')
    mlflow.set_tag('model_type','descision tree')
    model_tree.fit(train_tfidf,train_label)
    mlflow.log_params(model_tree.get_params())
    model_evaluate(model_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier
with mlflow.start_run(run_name='Random Forest v5'):
    model_forest=RandomForestClassifier(n_jobs=-1,n_estimators=120,max_depth=250,random_state=42,class_weight='balanced')
    model_forest.fit(train_tfidf,train_label)
    mlflow.set_tag('model_type','random forest')
    mlflow.log_params(model_forest.get_params())
    model_evaluate(model_forest)
    
```

```python
from sklearn.ensemble import RandomForestClassifier
model_forest=RandomForestClassifier(n_jobs=-1,n_estimators=120,max_depth=175,random_state=42,class_weight='balanced')
model_forest.fit(train_tfidf,train_label)
model_evaluate(model_forest)
```

```python
train_inputs=train_set['text'].to_numpy()
val_inputs=val_set['text'].to_numpy()
train_target=train_set['code'].to_numpy() #train_label are integers but pd.Series but train_target is an array of integres
val_target=val_set['code'].to_numpy()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization,Embedding,Dense
```

```python
text_vectorizer=TextVectorization(max_tokens=None,ngrams=None,output_mode='int',split='whitespace',
                                  standardize='lower_and_strip_punctuation',output_sequence_length=None)
```

```python
# sum([len(i.split()) for i in train_inputs])/len(train_inputs)
max_vocab=10000
tl=[len(i.split()) for i in train_inputs]
print(np.percentile(tl,[50,75,90,95,99]))
max_length=42
```

```python
text_vectorizer=TextVectorization(max_tokens=max_vocab,output_sequence_length=max_length)
```

```python
text_vectorizer.adapt(train_inputs)
```

```python
embed=Embedding(input_dim=max_vocab,output_dim=128)
```

```python
train_set['sentiment'].unique()
```

```python
from tensorflow.keras.layers import Dense,GRU,LSTM,Bidirectional,Input,Dropout
from tensorflow.keras.losses import CategoricalCrossentropy,SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
```

```python
def model_evaluate_dl(model):
    train_preds=np.argmax(model.predict(train_inputs),axis=1)
    val_preds=np.argmax(model.predict(val_inputs),axis=1)
    train_acc=accuracy_score(train_target,train_preds)
    val_acc=accuracy_score(val_target,val_preds)
    train_f1=f1_score(train_target,train_preds,average='weighted')
    val_f1=f1_score(val_target,val_preds,average='weighted')
    print(f'Train Results: Accuracy ==> {train_acc} and F1 ==> {train_f1}')
    print(f'Test Results: Accuracy ==> {val_acc} and F1 ==> {val_f1}')
    mlflow.log_metric('Train_Accuracy',train_acc)
    mlflow.log_metric('Train_F1',train_f1)
    mlflow.log_metric('Test_Accuracy',val_acc)
    mlflow.log_metric('Test_F1',val_f1)
```

```python
with mlflow.start_run(run_name='GRU v1'):
    tf.random.set_seed(42)
    inputs=Input(shape=(1,),dtype=tf.string)
    x=text_vectorizer(inputs)
    x=embed(x)
# x=GRU(32,activation='tanh',return_sequences=True,dropout=0.2,recurrent_dropout=0.2)(x)
    x=GRU(32,activation='tanh',dropout=0.3,recurrent_dropout=0.3,return_sequences=True)(x)
    x=GRU(32,activation='tanh',dropout=0.3,recurrent_dropout=0.3)(x)
    x=Dropout(0.3)(x)
    outputs=Dense(4,activation='softmax')(x)
    model_1=Model(inputs,outputs)
    model_1.compile(loss=SparseCategoricalCrossentropy(),optimizer=Adam(),metrics=['accuracy'])
    early_stopping=EarlyStopping(
       monitor='val_loss',patience=2,restore_best_weights=True
     )
    model_1.fit(train_inputs,train_target,epochs=25,validation_data=(val_inputs,val_target),callbacks=[early_stopping],batch_size=64)

    mlflow.log_params({'epochs':25,'model_type':'GRU','optimizer':'Adam','lr':0.001,'callbacks':'early_stopping','layers':2,
                       'units_per_layer':'[32,32]','dropout':0.3,'recurrent_dropout':0.3,'loss':'SparseCategoricalCrossentropy','activation':'tanh'})
    model_evaluate_dl(model_1)
    mlflow.set_tag('model_type','GRU')
```

```python
with mlflow.start_run(run_name='LSTM v1'):
    inputs=Input(shape=(1,),dtype=tf.string)
    x=text_vectorizer(inputs)
    x=embed(x)
    x=LSTM(32,activation='tanh',return_sequences=True,recurrent_dropout=0.2,dropout=0.3)(x)
    x=LSTM(32,activation='tanh',recurrent_dropout=0.2,dropout=0.3)(x)
    x=Dropout(0.3)(x)
    outputs=Dense(4,activation='softmax')(x)
    model_2=Model(inputs,outputs)
    model_2.compile(loss=SparseCategoricalCrossentropy(),optimizer=Adam(),metrics=['accuracy'])
    early_stopping=EarlyStopping(
    monitor='val_loss',patience=2,restore_best_weights=True
    )
    model_2.fit(train_inputs,train_target,epochs=20,validation_data=(val_inputs,val_target),callbacks=[early_stopping],batch_size=64)
    mlflow.log_params({'epochs':20,'model_type':'LSTM','optimizer':'Adam','lr':0.001,'callbacks':'early_stopping','layers':2,
                       'units_per_layer':'[32,32]','dropout':0.3,'recurrent_dropout':0.2,'loss':'SparseCategoricalCrossentropy','activation':'tanh'})
    model_evaluate_dl(model_2)
    mlflow.set_tag('model_type','LSTM')    
```

```python
with mlflow.start_run(run_name='GRU v2'):
    tf.random.set_seed(42)
    inputs=Input(shape=(1,),dtype=tf.string)    
    x=text_vectorizer(inputs)
    x=embed(x)
# x=GRU(32,activation='tanh',return_sequences=True,dropout=0.2,recurrent_dropout=0.2)(x)
    x=GRU(32,activation='tanh',dropout=0.3,recurrent_dropout=0.3,return_sequences=True)(x)
    x=GRU(32,activation='tanh',dropout=0.3,recurrent_dropout=0.3)(x)
    x=Dropout(0.3)(x)
    outputs=Dense(4,activation='softmax')(x)
    model_5=Model(inputs,outputs)
    model_5.compile(loss=SparseCategoricalCrossentropy(),optimizer=Adam(),metrics=['accuracy'])
    reduce_lr=ReduceLROnPlateau(patience=2,monitor='val_loss',factor=0.5,min_lr=1e-6)
    early_stopping=EarlyStopping(
       monitor='val_loss',patience=2,restore_best_weights=True
     )
    model_5.fit(train_inputs,train_target,epochs=25,validation_data=(val_inputs,val_target),callbacks=[early_stopping,reduce_lr],batch_size=64)

    mlflow.log_params({'epochs':25,'model_type':'GRU','optimizer':'Adam','lr':0.001,'callbacks':'early_stopping , ReduceLROnPlateau','layers':2,
                       'units_per_layer':'[32,32]','dropout':0.3,'recurrent_dropout':0.3,'loss':'SparseCategoricalCrossentropy','activation':'tanh'})
    mlflow.set_tag('modification','ReduceLr callback is used on gru v1 to see any performance difference')
    model_evaluate_dl(model_5)
    mlflow.set_tag('model_type','GRU')
```
